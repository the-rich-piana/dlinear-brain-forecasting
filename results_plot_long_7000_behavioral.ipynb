```python
def plot_active_vs_passive_comparison(active_results, passive_results, ax, y_max=None):
    """Plot model performance comparison between Active and Passive behavioral data"""
    
    # Get models excluding Naive for bar plot
    models_to_plot = ['Mean', 'POCO', 'TSMixer', 'Linear', 'DLinear', 'Informer', 'Transformer']
    
    # Prepare data for side-by-side bars
    active_data = active_results[active_results['model'].isin(models_to_plot)].copy()
    passive_data = passive_results[passive_results['model'].isin(models_to_plot)].copy()
    
    # Add dataset labels
    active_data['dataset'] = 'Active'
    passive_data['dataset'] = 'Passive'
    
    # Combine data
    combined_data = pd.concat([active_data, passive_data], ignore_index=True)
    
    # Define colors - brighter green for Active, purpley blue for Passive
    colors = {'Active': '#32CD32', 'Passive': '#6A5ACD'}  # Lime Green and Slate Blue
    
    # Create grouped bar plot
    sns.barplot(data=combined_data, x='model', y='mae', hue='dataset', 
                palette=colors, ax=ax)
    
    # Add Naive reference lines with matching colors
    active_naive_mae = active_results[active_results['model'] == 'Naive']['mae'].iloc[0]
    passive_naive_mae = passive_results[passive_results['model'] == 'Naive']['mae'].iloc[0]
    
    # Create custom legend entries for Naive lines (don't add to plot yet)
    active_naive_line = plt.Line2D([0], [0], color=colors['Active'], linestyle='--', 
                                   linewidth=2, alpha=0.8)
    passive_naive_line = plt.Line2D([0], [0], color=colors['Passive'], linestyle='--', 
                                    linewidth=2, alpha=0.8)
    
    # Add the actual lines to the plot
    ax.axhline(y=active_naive_mae, color=colors['Active'], linestyle='--', 
               linewidth=2, alpha=0.8)
    ax.axhline(y=passive_naive_mae, color=colors['Passive'], linestyle='--', 
               linewidth=2, alpha=0.8)
    
    # Set consistent y-axis limits
    if y_max is not None:
        ax.set_ylim(0, y_max * 1.1)  # Add 10% padding
    
    # Styling
    ax.set_title('Model Performance: Active vs Passive Behavioral Data (7000 Neurons)', 
                 fontsize=14, fontweight='bold', pad=15)
    ax.set_xlabel('Model', fontsize=11, fontweight='bold')
    ax.set_ylabel('Mean Absolute Error (MAE)', fontsize=11, fontweight='bold')
    ax.tick_params(axis='x', rotation=45, labelsize=10)
    ax.grid(axis='y', alpha=0.3)
    
    # Create legend with bars + custom Naive lines
    bar_handles, bar_labels = ax.get_legend_handles_labels()
    
    # Combine bar handles with custom Naive line handles
    all_handles = bar_handles + [active_naive_line, passive_naive_line]
    all_labels = bar_labels + [f'Active Naive (MAE: {active_naive_mae:.4f})', 
                               f'Passive Naive (MAE: {passive_naive_mae:.4f})']
    
    ax.legend(all_handles, all_labels, fontsize=9, loc='upper right')
    
    # Add value labels on bars
    for container in ax.containers:
        ax.bar_label(container, fmt='%.4f', fontsize=8, rotation=90, padding=3)
    
    return ax

# Find the maximum MAE across both datasets to set consistent y-axis
all_mae_values = list(results7000_behavorial['mae']) + list(results7000['mae'])
max_mae = max(all_mae_values)
print(f"Using y-axis maximum: {max_mae:.4f}")

# Create single plot for comparison
fig, ax = plt.subplots(1, 1, figsize=(14, 8))

# results7000 is Active test data, results7000_behavorial is Passive test data
plot_active_vs_passive_comparison(results7000, results7000_behavorial, ax, y_max=max_mae)

plt.tight_layout()
plt.show()

# Print summary comparison table
print("\nSummary: Active vs Passive Behavioral Data Performance (MAE)")
print("=" * 60)
comparison_summary = pd.DataFrame({
    'Active_MAE': results7000.set_index('model')['mae'],
    'Passive_MAE': results7000_behavorial.set_index('model')['mae'].reindex(results7000['model'].values)
})
comparison_summary['Difference'] = comparison_summary['Active_MAE'] - comparison_summary['Passive_MAE']
comparison_summary['Improvement_in_Active'] = comparison_summary['Difference'] < 0
print(comparison_summary.round(4))
```

```python
# Function to calculate performance metrics relative to Naive baseline
def calculate_performance_metrics(results_df, dataset_name):
    """
    Calculate three performance metrics relative to Naive baseline:
    1. MAE improvement %: ((Naive_MAE - Model_MAE) / Naive_MAE)  100
    2. MSE improvement %: ((Naive_MSE - Model_MSE) / Naive_MSE)  100  
    3. Prediction Score: 1  L(model)/L(naive) where L is MSE loss
    """
    
    # Get Naive baseline values
    naive_row = results_df[results_df['model'] == 'Naive']
    if len(naive_row) == 0:
        print(f"Warning: No Naive model found in {dataset_name}")
        return None
        
    naive_mae = naive_row['mae'].iloc[0]
    naive_mse = naive_row['mse'].iloc[0]
    
    # Calculate metrics for all models
    metrics_df = results_df.copy()
    
    # 1. MAE improvement percentage
    metrics_df['mae_improvement_pct'] = ((naive_mae - metrics_df['mae']) / naive_mae) * 100
    
    # 2. MSE improvement percentage  
    metrics_df['mse_improvement_pct'] = ((naive_mse - metrics_df['mse']) / naive_mse) * 100
    
    # 3. Prediction Score: 1 - L(model)/L(naive)
    metrics_df['prediction_score'] = 1 - (metrics_df['mse'] / naive_mse)
    
    # Add dataset identifier
    metrics_df['dataset'] = dataset_name
    
    return metrics_df

print("Calculating performance metrics for Active vs Passive datasets...")
print("=" * 60)

# Calculate metrics for both datasets
active_metrics = calculate_performance_metrics(results7000, "Active")
passive_metrics = calculate_performance_metrics(results7000_behavorial, "Passive")

print("Metrics calculated successfully!")
print("\nFormulas used:")
print("1. MAE Improvement %: ((Naive_MAE - Model_MAE) / Naive_MAE)  100")
print("2. MSE Improvement %: ((Naive_MSE - Model_MSE) / Naive_MSE)  100") 
print("3. Prediction Score: 1  L(model)/L(naive) where L = MSE loss")
```

```python
# Create performance tables for Passive, Active, and Comparison
def create_performance_table(metrics_df, title):
    """Create formatted performance table"""
    # Exclude Naive from display (it's always 0% improvement, score=0)
    display_df = metrics_df[metrics_df['model'] != 'Naive'].copy()
    
    # Select and rename columns for display
    table_df = display_df[['model', 'mae_improvement_pct', 'mse_improvement_pct', 'prediction_score']].copy()
    table_df.columns = ['Model', 'MAE Improv %', 'MSE Improv %', 'Pred Score']
    
    # Round values for readability
    table_df['MAE Improv %'] = table_df['MAE Improv %'].round(2)
    table_df['MSE Improv %'] = table_df['MSE Improv %'].round(2)
    table_df['Pred Score'] = table_df['Pred Score'].round(4)
    
    print(f"\n{title}")
    print("-" * 70)
    print(table_df.to_string(index=False))
    
    return table_df

print("\nACTIVE VS PASSIVE BEHAVIORAL DATA PERFORMANCE METRICS")
print("=" * 70)
print("All metrics calculated relative to Naive baseline within each dataset")
print("=" * 70)

# 1. Passive dataset table
passive_table = create_performance_table(passive_metrics, "PASSIVE Behavioral Data Performance")

# 2. Active dataset table  
active_table = create_performance_table(active_metrics, "ACTIVE Behavioral Data Performance")

# 3. Comparison table showing differences (Active - Passive)
print(f"\n\nCOMPARISON: Changes in Performance Metrics (Active - Passive)")
print("-" * 70)
print("Positive values = Better performance in Active dataset")

# Merge on model to compare metrics
comparison_df = active_metrics.merge(passive_metrics, on='model', suffixes=('_active', '_passive'))

comparison_table = pd.DataFrame({
    'Model': comparison_df['model'],
    'MAE Improv ': (comparison_df['mae_improvement_pct_active'] - comparison_df['mae_improvement_pct_passive']).round(2),
    'MSE Improv ': (comparison_df['mse_improvement_pct_active'] - comparison_df['mse_improvement_pct_passive']).round(2),
    'Pred Score ': (comparison_df['prediction_score_active'] - comparison_df['prediction_score_passive']).round(4)
})

# Exclude Naive from comparison (always 0)
comparison_display = comparison_table[comparison_table['Model'] != 'Naive']
print(comparison_display.to_string(index=False))

print(f"\n\nNAIVE BASELINES (Reference Values)")
print("-" * 50)
baseline_comparison = pd.DataFrame({
    'Dataset': ['Active', 'Passive'],
    'Naive MAE': [active_metrics.iloc[0]['mae'], passive_metrics.iloc[0]['mae']],
    'Naive MSE': [active_metrics.iloc[0]['mse'], passive_metrics.iloc[0]['mse']]
})
baseline_comparison['Naive MAE'] = baseline_comparison['Naive MAE'].round(4)
baseline_comparison['Naive MSE'] = baseline_comparison['Naive MSE'].round(4)
print(baseline_comparison.to_string(index=False))
```
